```python
# Dictionary (딕셔너리): 자바스크립트의 객체

# dictionary 데이터 생성
dic1 = {} # empty dictionary
dic2 = {
  '학번': 3,
  '이름': 'kim',
  '점수': 80
}

print(dic2)
print(dic2['학번'])
```

```python
# 데이터 추가 및 변경
phone = {
  'modelName': 's20',
  'price': 1000
}

# 데이터 추가
# 없는 key에 값을 저장하면 새로운 key가 추가됨
phone['color'] = 'red' # color라는 key를 생성하고 value에 'red' 문자열을 저장
print(phone)

# 데이터 수정
phone['price'] = 2000
print(phone)
```

```python
# dictionary 데이터 삭제 del 명령어 사용
del phone['color']
print(phone)
```
