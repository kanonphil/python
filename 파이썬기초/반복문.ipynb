```python
# 파이썬은 for-each 문만 있음

list1 = [2, 4, 6, 8, 10]
for e in list1:
  print(e)
```

```python
# for-each 문을 일반 for문처럼 사용하기
# for(int i = 0; i < 10; i++)
for e in range(1, 10):
  print(e)

for e in range(2, 5):
  print(e)

for e in range(10): # 0부터 10 전까지
  print(e)
```

```python
# 반복문의 index 정보 알아내기
# 반복문의 index를 알고 싶다면 enumerate() 함수 매개변수에 리스트를 넣어 사용
list1 = [2, 4, 6, 8, 10]
for e, i in enumerate(list1): # e: 데이터 하나하나, i: index
  print(f'데이터: {e}, 인덱스: {i}')
```
