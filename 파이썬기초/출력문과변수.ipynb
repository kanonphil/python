파이썬의 출력문

```python
# 파이썬 주석
# 파이썬은 명령어 끝에 세미콜론(;) 안 붙임!
print(1)
print('문자열')
print("문자열")
print(10 + 20)
print('abc' + 'def')
# print('abc' + '1') 오류! 문자열과 숫자 더하기는 불가능
```

```python
# 연산자
print(3 + 5)
print(3 * 5)
print(10 / 4) # 2.5
print(10 // 4) # 2 (몫 구하기)
print(10 % 4) # 나머지 계산
```

파이썬의 변수 사용

```python
# 파이썬 변수는 내부적으로만 자료형이 존재
# 변수 선언 시 자료형을 작성하지 않는다.

# 이름을 저장할 수 있는 선언 및 초기화
myName = 'hong'
print(myName)

myAge = 20
myHeight = 170.5

# type(변수): 매개변수로 전달된 데이터의 자료형을 확인
print(type(myAge))
print(type(myHeight))
```

```python
# 문자열과 변수를 함께 출력하는 방법
addr = '울산시'
score = 90

# str(): 매개변수로 전달된 숫자를 문자열로 반환
# int(): 매개변수로 전달된 문자열을 정수로 반환
# '사는 곳은 울산시이며, 제 점수는 90점 입니다.'
print(f'사는 곳은 {addr}이며, 제 점수는 {score}점 입니다.')
```
