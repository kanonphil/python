```python
# 리스트 (다수의 데이터를 index 번호로 관리하는 자료형)
# 변수명에 list는 사용 금지!
list1 = [] # 빈 리스트 생성
list2 = [1, 5.5, 'python'] # 서로 다른 자료형 저장 가능

print(list1)
print(list2)
print(list2[1])
```

```python
# 리스트에 데이터 추가하기
list1 = [1, 2, 3]
list1.append(4)
list1.insert(5, 5)
list1.extend(list2)
print(list1)

# 1. append(data): 자바스크립트의 push와 같음
# 2. insert(index, data): 원하는 index에 데이터를 추가
```

```python
# 리스트 데이터 수정
list2 = [1, 3.3, 'python']

# 1번 요소를 'hello'로 변경
list2[1] = 'hello'

print(list2)
```

```python
# 리스트 합치기
list1 = [1, 2, 3]
list2 = [4, 5, 6]

list3 = list1 + list2
print(list3)

list4 = list1 * 3
print(list4)
```

```python
# 리스트의 슬라이싱 연산 (리스트의 특정 부분만 추출)
# [start : end] end 포함 안함
list1 = [1, 2, 3, 4, 5]
print(list1[1:4])

list2 = list1[1:3]
print(list2)

print(list4[0:9:3])
```

```python
# python에서는 문자열을 리스트로 취급
myName = 'hong'
print(myName[1])

addrList = [
  '스타벅스 삼산지점 052-1111-2222',
  '스타벅스 신정동 2호점 052-2222-3333',
  '스타벅스 울산남구점 052-3333-4444',
]

# 스타벅스 매장 정보에서 연락처 정보만 추출
print(addrList[0][-13:])
print(addrList[1][-13:])
print(addrList[2][-13:])

for i in range(len(addrList)):
  print(addrList[i][-13:])
```
