```python
# 리턴이 있는 함수 + 매개변수가 있는 함수
def getSum(a, b):
  return a + b

# ipynb 파일은 마지막에 값을 실행하면 자동 출력
getSum(10, 20)
```

```python
# 파이썬은 리턴 데이터를 여러 개 가질 수 있음
def getNumber():
  return 10, 20

a, b = getNumber()
print(a)
print(b)
```

```python
# 파이썬은 매개변수 사용 시 매개변수 이름을 자주한다
def tellMe(name, age, addr):
  print(f'이름은 {name}, 나이는 {age}, 주소는 {addr}')
```

```python
tellMe('kim', 20, 'ulsan')
tellMe(name='kim', age=20, addr='ulsan')
tellMe(age=20, addr='ulsan', name='kim')
```

```python
def tellMe2(name='미정', age=1, addr='알 수 없음'):
  print(f'이름은 {name}, 나이는 {age}, 주소는 {addr}')
```
