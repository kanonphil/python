```python
num1 = 2

# 파이썬 조건문의 실행 영역은 들여쓰기로 범위를 구분한다.
if num1 > 3 : 
  print('num1은 3보다 큽니다')
else :
  print('num1은 3보다 작습니다.')
```

```python
num2 = 3

# num2 변수의 값이 3보다 크면 1, 3보다 작으면 2, 3이면 0 출력
if num2 > 3 :
  print(1)
elif num2 < 3 :
  print(2)
else :
  print(0)
```

```python
# input(): 키보드로 데이터 입력받기
inputedData = input('정수 입력: ')
print(inputedData)
```

```python
# 키보드로 입력받은 점수가 짝수인지 홀수인지 판별 및 출력
inputScore = int(input('점수 입력: '))
if inputScore % 2 == 0 :
  print('짝수')
else :
  print('홀수')
```

```python
# 키보드로 입력받은 단수의 구구단을 출력
inputNum = int(input('구구단 단 수 입력: '))
for i in range(1, 10) :
  print(f'{inputNum} * {i} = {inputNum * i}')
```
