```python
# 주어진 딕셔너리 데이터로 1~2번 문제를 푸세요.
student = {
  'name': 'hong',
  'korScore': 0,
  'mathScore': 0
}
```

```python
# 1. 위 딕셔너리 데이터에 키보드로 입력한 값으로 국어점수 및 수학점수를 변경
score1 = int(input('국어점수: '))
score2 = int(input('수학점수: '))

student['korScore'] = score1
student['mathScore'] = score2

print(student)
```

```python
# 2. 위 딕셔너리 데이터에 총점(totalScore)을 추가하고, value에는 국어 및 수학 점수의 합을 저장
total = student['korScore'] + student['mathScore']

student['totalScore'] = total

print(student)
```

```python
# 3. 아래 딕셔너리 데이터가 매개변수로 주어지면 모든 점수의 합을 리턴하는 함수 선언 및 호출
scores = {
  'kor': 70,
  'math': 80,
  'eng': 90
}

# 함수 선언
def getSum(dic):
  return sum(dic.values(), 0)
```

```python
# 3번 문제의 함수를 호출
getSum(scores)
```

```python
'''
매개변수로 전달된 딕셔너리에 주어진 key와 value를 추가한 딕셔너리를 리턴하는 함수를 선언 및 호출
첫 번째 매개변수: 딕셔너리
두 번째 매개변수: 추가할 key
세 번째 매개변수: 추가할 value
'''

# 4번 함수 선언
def addDict(dictionary1, key, value):
  dictionary1[key] = value
  return dictionary1
```

```python
# 4번 함수 호출
my_dict = {'name': 'kim', "age": 25}
addDict(my_dict, 'addr', 'ulsan')
print(my_dict)
```

```python
'''
5번
매개변수로 문자열 리스트가 전달되면 리스트이 문자열과 문자열의 길이를 딕셔너리로 리턴하는 함수 선언 및 호출

ex> 매개변수: ['apple', 'banana', 'orange']
결과: {
  'apple': 5,
  'banana': 6,
  'orange': 6
}
'''

# 5번 함수 선언
def listToDict(str_list):
  result = {}
  for word in str_list:
    result[word] = len(word)

  return result
```

```python
# 5번 함수 호출
list1 = ['apple', 'banana', 'orange']
print(listToDict(list1))
```
