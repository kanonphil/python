```python
# 결측치 확인
df.isnull() # 각 데이터를 True, False로 결측치 표현

# 결측치가 있는 행만 보기
# axis=0 -> 행, axis=1 -> 열
df[df.isnull().any(axis=1)]

# 각 컬럼별 결측치 갯수 확인
df.isnull().sum()
```

```python
# df의 내용을 test_df로 복사
test_df = df.copy()
test_df

```

```python
# 결측치 제거
# dropna(axis='' how='')
# axis='index' 결측치가 있는 행 삭제 (default)
# axis='columns' 결측치가 있는 열 삭제 (axis=1과 같음)
# how='any' 결측치가 하나라도 존재하면 삭제 (default)
# how='all' 모든 데이터가 결측치 일 때 삭제

# dropna() 함수를 사용하면 결측치가 제거된 결과 데이터를 리턴, 원본 데이터는 유지
result1_df = test_df.dropna()

# 결측치가 제거된 데이터
result1_df

# test_df를 출력하면 원본 데이터 유지
test_df

# 컬럼 기준으로 결측치가 1개라도 존재하는 열을 삭제
result2_df = test_df.dropna(axis=1, how='any')
result2_df

# 특정 컬럼에 결측치가 존재하는 행만 제거
# ex> 파이썬 컬럼에 결측치가 존재하는 행 제거
result3_df = test_df.dropna(subset=['파이썬점수']) # 파이썬컬럼에 결측치가 존재하는 행 제거
result3_df

# ex> 리스트에 나열한 컬럼 중 하나라도 결측치가 포함되어 있으면 삭제
result4_df = test_df.dropna(subset=['파이썬점수', '자바점수'])
result4_df
```

```python
# 6. 아래 요구사항에 맞게 결측치를 채우세요.
#  1) 부서의 결측치는 '미배정'으로 채움
#  2) 연봉의 결측치는 연봉 평균으로 채움
#  3) 성과점수 결측치는 중앙값으로 채움
r_df3 = t_df.fillna({
  '부서': '미배정',
  '연봉': t_df['연봉'].mean(),
  '성과점수': t_df['성과점수'].median()
})

r_df3
```
